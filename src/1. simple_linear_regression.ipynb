```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data={
    "Area": [ 2600, 3000, 3200, 3600, 4000],
    "Price": [ 550000, 565000, 610000, 680000,725000]
}
df= pd.DataFrame(data)
df.head()
```

```python
%matplotlib inline
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area~Price")
plt.scatter(df['Area'],df['Price'], color= 'black', marker= '*')
plt.gca().set_facecolor("silver")
plt.show()
```

```python
slr= linear_model.LinearRegression()
slr.fit(df[["Area"]],df['Price'])
```

```python
predict={
    "Area": [3100,3300]
}
predict_price= slr.predict(pd.DataFrame(predict))
print(predict_price)
print(int(predict_price[0]))
print(int(predict_price[1]))
```
